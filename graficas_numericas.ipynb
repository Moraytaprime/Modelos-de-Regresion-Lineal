```python
#Carga desde un archivo .csv sin indice y con decodificación. Ejemplos: 'ascii', 'UTF-8'
df = pd.read_csv("DataframeEquipo3Procesado.csv")
df.head(30)
```

```python
df.iloc[:,[28]]
```

```python
df['Equipo del entrevistador'] = df['Equipo del entrevistador'].astype(str)
df['¿Cuántos clientes, en promedio, recibe en un día entre semana?'] = df['¿Cuántos clientes, en promedio, recibe en un día entre semana?'].astype(str)
df['¿Cuántos clientes, en promedio, recibe en un día del fin de semana?'] = df['¿Cuántos clientes, en promedio, recibe en un día del fin de semana?'].astype(str)
df['¿Cuántos productos aproximadamente maneja en su tienda?'] = df['¿Cuántos productos aproximadamente maneja en su tienda?'].astype(str)
df['En caso de tener un sistema de recolección ¿Cada cuánto pasa el camión de la basura por su negocio?'] = df['En caso de tener un sistema de recolección ¿Cada cuánto pasa el camión de la basura por su negocio?'].astype(str)
df['En una escala de 1 a 10, ¿cómo evaluaría sus esfuerzos para reducir su impacto ambiental?'] = df['En una escala de 1 a 10, ¿cómo evaluaría sus esfuerzos para reducir su impacto ambiental?'].astype(str)
df['En una escala de 1 a 10, ¿cómo evaluaría sus esfuerzos para reducir su impacto social?'] = df['En una escala de 1 a 10, ¿cómo evaluaría sus esfuerzos para reducir su impacto social?'].astype(str)

```

```python

#Obtengo un análisis univariado de una variable categórica en específico
table1= freq_tbl(df.iloc[:,[28]])

#Obtengo un filtro de los valores más reelevantes de la variables categórica seleccionada
Filtro= table1[table1['frequency']>1]
#Ajusto el indice de mi dataframe
Filtro_index= Filtro.set_index('Equipo del entrevistador')

#Realizamos grafico de pastel del dataframe filtrado
Filtro_index["frequency"].plot(kind='pie', figsize=(10,5), shadow=True, autopct="%0.1f %%")
```
